```python
new_dataset = new_dataset.drop('Observation', axis=1)
```

```python
new_dataset = new_dataset.drop_duplicates()

# Convert dataset columns to numeric data types

new_dataset = new_dataset.apply(pd.to_numeric, errors='coerce')

q1 = new_dataset.quantile(0.25)
q3 = new_dataset.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

# Outliers Data
outliers = new_dataset[(new_dataset < lower_bound) | (new_dataset > upper_bound)]

# Data with no outliers
no_outliers = new_dataset[(new_dataset >lower_bound) & (new_dataset <upper_bound)]


no_outliers
```

```python
no_outliers.dropna()
```

```python
#backfilling and forward filling the null values
no_outliers=no_outliers.bfill()
no_outliers=no_outliers.ffill()
no_outliers
```
